# Historical Pathways: Prompt Engineering and Design

Historical Pathways is a conceptual idea for an app that enables users to explore how historical events are connected. 

This notebook looks at how prompts about historical events might write their own follow-up prompts. The goal is to explore the viability of an app that guides users through open-ended, yet coherent historical narratives. The core concept is to be able to create narratives from any given starting point in history and enable an experience that is partly AI-guided, and partly human-guided. 

### Prompt Types

- System prompt
- Event prompt
- People prompt
- Place prompt

### Overall flow of prompt calls

- Step 1 (event prompt):
    - Determine if historical event is too broad, give related events that will work.
    - Get Summary, Introduction (introduce event, why it is important, any major figures), Themes, People, Places.
- Step 2 (from event prompt):
    - User’s choose where to go next
    - OR auto-play: Users can influence auto-play by choosing
        - Broad themes to explore (e.g., "Emergence of Political Parties", "Social Justice")
